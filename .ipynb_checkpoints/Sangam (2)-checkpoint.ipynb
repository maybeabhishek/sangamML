```python
import numpy as np
import pandas as pd
import datetime
trainDF = pd.read_csv("./datasets/Train.csv")
testDF = pd.read_csv("./datasets/Test.csv")
trainDF.shape
```

```python
time=[]
for i in trainDF['date_time']:
    time.append(i[11:])
trainDF['time']=pd.DataFrame(time)    
trainDF['time'].head()
```

```python
time=[]
for i in trainDF['date_time']:
    time.append(i[:4])
trainDF['year']= pd.DataFrame(time)

time=[]
for i in trainDF['date_time']:
    time.append(i[5:7])
trainDF['month']= pd.DataFrame(time)    

time=[]
for i in trainDF['date_time']:
    time.append(i[8:10])
trainDF['date']=pd.DataFrame(time)    

day=[]
for i in range(len(trainDF)):
    temp=datetime.date(int(trainDF['year'][i]),int(trainDF['month'][i]),int(trainDF['date'][i]))
    day.append(datetime.date.weekday(temp))
trainDF['weekday']=pd.DataFrame(day)

trainDF.head()
```

```python
trainDF['is_holiday_label'] = np.where(trainDF['is_holiday'] == 'None', 0,1)
trainDF.head()
```

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib

numericalColumns = np.delete(columns, [0, 1, 12, 13, 15, 16, 17, 18, 19, 20])
numericalColumns
```

```python
matplotlib.rcParams['figure.figsize'] = (15, 15)
for i in range(0, len(numericalColumns)):
    plt.subplot(5, 4, i+1)
    sns.distplot(trainDF[numericalColumns[i]])
    plt.xticks([])
    plt.yticks([])
```

- rain_p_h and sno_p_h are very sparse and can be dropped
- visibility and dew point are perfectly categorical

```python
numericalColumns = numericalColumns[numericalColumns!='rain_p_h']
numericalColumns = numericalColumns[numericalColumns!='snow_p_h']
numericalColumns
```

```python
matplotlib.rcParams['figure.figsize'] = (15, 30)
for i in range(0, len(numericalColumns)):
    plt.subplot(5, 4, i+1)
    sns.boxplot(trainDF[numericalColumns[i]], orient='v')
    plt.xticks([])
    plt.yticks([])
```

## Remove outliars

```python
from scipy import stats
trainDF = trainDF[(np.abs(stats.zscore(trainDF[numericalColumns])) < 3).all(axis=1)]
for i in range(0, len(numericalColumns)):
    plt.subplot(5, 4, i+1)
    sns.boxplot(trainDF[numericalColumns[i]], orient='v')
    plt.xticks([])
    plt.yticks([])
```

```python
categoricalColumns = np.setdiff1d(columns, numericalColumns)
categoricalColumns = categoricalColumns[categoricalColumns!='rain_p_h']
categoricalColumns = categoricalColumns[categoricalColumns!='snow_p_h']
categoricalColumns
```

```python
trainDF[categoricalColumns] = trainDF[categoricalColumns].astype('category')
trainDF.info()
```

```python
categoricalDF= trainDF.copy(deep=True)
for i in weather_type:
    categoricalDF[i]=0
categoricalDF.shape
```

```python
for index, row in categoricalDF.iterrows():
    for i in weather_type:
        if i.lower() in row['weather_description'].lower().split(' '):
            categoricalDF.at[index,i]=1

categoricalDF[weather_type].head()
```

```python
try:
    categoricalDF = categoricalDF.drop(['weather_type', 'weather_description','traffic_volume','rain_p_h', 'snow_p_h', 'is_holiday'], axis=1)
except KeyError:
    pass
categoricalDF.columns.shape
```

```python
numericalColumns = numericalColumns[numericalColumns!='traffic_volume']
numericalColumns.shape
```

```python
scaledDF = categoricalDF.copy(deep=True)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaledDF[numericalColumns] = scaler.fit_transform(scaledDF[numericalColumns])
scaledDF = scaledDF.drop(['date_time'], axis=1)
```

```python
scaledDF.head()
```

```python
scaledDF.columns
```

```python
scaledDF = scaledDF.drop(['time', 'month', 'year', 'date', 'weekday'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X = scaledDF.values
y = trainDF['traffic_volume'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.svm import SVC
svm_clf = SVC(kernel="linear", C=float("inf"),verbose=1)
svm_clf.fit(X_train, y_train)
```

```python
svm_clf.predict(X_test,y_test)
```
